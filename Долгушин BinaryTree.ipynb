```python
class BinaryTree():
  def __init__(self, root=None):
    self.root = root
  
  def add_node(self, node):
    if self.root is None:
      self.root = node
    else:
      current = self.root
      while current.key != node.key:
        if current.key < node.key and current.right is not None:
          current = current.right
        elif current.key > node.key and current.left is not None:
          current = current.left
        else:
          if current.key < node.key:
            current.right = node
          elif current.key > node.key:
            current.left = node
          node.parent = current
          break
      else:
        raise Exception("Key is UNACCEPTABLE! (the key already exists)")
  
  def traverse_tree(self):
    for i in self.root.show():
      yield i

  def __getitem__(self, key):
    if self.root is None:
      raise Exception("Key is UNACCEPTABLE! (tree is empty)")
    else:
      current = self.root
      while current.key != key:
        if current.key < key and current.right is not None:
          current = current.right
        elif current.key > key and current.left is not None:
          current = current.left
        else:
          raise Exception("Key is UNACCEPTABLE! (no such key)")
      return current
  
  def remove(self, key):
    wannabe_removed = self[key]
    if wannabe_removed.right is None and wannabe_removed.left is None:
      if wannabe_removed.parent is not None and wannabe_removed.parent.right == wannabe_removed:
        wannabe_removed.parent.right = None
      elif wannabe_removed.parent is not None and wannabe_removed.parent.left== wannabe_removed:
        wannabe_removed.parent.left = None
      elif wannabe_removed.parent is None:
          self.root = None
    elif wannabe_removed.left is None:
      if wannabe_removed.parent is not None and wannabe_removed.parent.right == wannabe_removed:
        wannabe_removed.parent.right = wannabe_removed.right
        wannabe_removed.right.parent = wannabe_removed.parent
      elif wannabe_removed.parent is not None and wannabe_removed.parent.left== wannabe_removed:
        wannabe_removed.parent.left = wannabe_removed.right
        wannabe_removed.right.parent = wannabe_removed.parent
      elif wannabe_removed.parent is None:
          self.root = wannabe_removed.right
    elif wannabe_removed.right is None:
      if wannabe_removed.parent is not None and wannabe_removed.parent.right == wannabe_removed:
        wannabe_removed.parent.right = wannabe_removed.left
        wannabe_removed.left.parent = wannabe_removed.parent
      elif wannabe_removed.parent is not None and wannabe_removed.parent.left== wannabe_removed:
        wannabe_removed.parent.left = wannabe_removed.left
        wannabe_removed.left.parent = wannabe_removed.parent
      elif wannabe_removed.parent is None:
          self.root = wannabe_removed.left
    else:
      if wannabe_removed.right.left is not None:
        current = wannabe_removed.right.left
        while current.left is not None:
          current = current.left
        current.parent.left = current.right
        if current.right is not None:
          current.right.parent = current.parent
        
        current.right = None
        current.parent = wannabe_removed.parent
        current.left = wannabe_removed.left
        current.right = wannabe_removed.right
        if wannabe_removed.parent is not None and wannabe_removed.parent.right == wannabe_removed:
          wannabe_removed.parent.right = current
        elif wannabe_removed.parent is not None and wannabe_removed.parent.left== wannabe_removed:
          wannabe_removed.parent.left = current
        elif wannabe_removed.parent is None:
          self.root = current
      else:
        wannabe_removed.right.left = wannabe_removed.left
        wannabe_removed.left.parent = wannabe_removed.right
        wannabe_removed.right.parent = wannabe_removed.parent
        if wannabe_removed.parent is not None and wannabe_removed.parent.right == wannabe_removed:
          wannabe_removed.parent.right = wannabe_removed.right
        elif wannabe_removed.parent is not None and wannabe_removed.parent.left== wannabe_removed:
          wannabe_removed.parent.left = wannabe_removed.right
        elif wannabe_removed.parent is None:
          self.root = wannabe_removed.right

```

```python
class Node():
  def __init__(self, key, value, parent=None):
    self.left = None
    self.right = None
    self.key = key
    self.value = value
    self.parent = parent
  
  def show(self):
    yield self.key, self.value
    if self.left is not None:
      for i in self.left.show():
        yield i
    if self.right is not None:
      for i in self.right.show():
        yield i
    
```
