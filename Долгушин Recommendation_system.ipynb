```python
def euqlid(v1, v2):
  k = 0
  for i in v1:
    if i in v2:
      k += (v1[i] - v2[i]) ** 2
  res = 1 / (1 + k)
  return res
```

```python
from math import sqrt

def pearson(sv1,sv2):
  v1 = []
  v2 = []
  for i in sv1:
    if i in sv2:
      v1 += [sv1[i]]
      v2 += [sv2[i]]
  sum1=sum(v1)
  sum2=sum(v2)

  sum1Sq=sum([pow(v,2) for v in v1])
  sum2Sq=sum([pow(v,2) for v in v2])	
  
  # Sum of the products

  pSum=sum([v1[i]*v2[i] for i in range(len(v1))])
  
  # Calculate r (Pearson score)
  length_of_set = len(v1)
  if len(v1) == 0:
    length_of_set = 10000000
  num=pSum-(sum1*sum2/length_of_set)
  den=sqrt((sum1Sq-pow(sum1,2)/length_of_set)*(sum2Sq-pow(sum2,2)/length_of_set))

  if den==0: return 0
  return 1.0-num/den
```

```python
def get_recom_table(db, distance="euqlid"):
  res = {}
  for i in db.keys():
    for k in db.keys():
      if i != k:
        if k + "_" + i not in res.keys():
          if distance == "euqlid":
            res[i + "_" + k] = euqlid(db[i], db[k])
          else:
            res[i + "_" + k] = pearson(db[i], db[k])
  return res
```

Вывести 5 ближайших

```python
def find_five_nearest(tbl):
  tbl = dict(sorted(tbl.items(), key=lambda item: -item[1]))
  i = 0
  for key, value in tbl.items():
    print(f"{key} : {value}")
    i += 1
    if i == 5:
      break
```
