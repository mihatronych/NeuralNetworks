```python
class GraphOnLists:
  def __init__(self):
    self.nodes = []
    self.node_edges = {}
  
  def add_node(self, node):
    if node in self.nodes:
      raise Exception("Вы пытаетесь добавить один и тот же узел дважды!")
    self.nodes.append(node)
    self.node_edges[node] = []
      
  def add_link(self, node1, node2):
    if node1 not in self.nodes or node2 not in self.nodes:
      raise Exception("Вы пытаетесь добавить связь с узлом, которого нет в графе!")
    if not (node2 not in self.node_edges[node1] and node1 not in self.node_edges[node2]):
      raise Exception("Вы пытаетесь добавить связь, которая уже есть в графе!")
    self.node_edges[node1].append(node2)
    self.node_edges[node2].append(node1)
      
  def remove_node(self, node):
    if node not in self.nodes:
      raise Exception("Вы пытаетесь удалить узел, которого нет!")
    self.nodes.remove(node)
    self.node_edges.pop(node)
    for k, v in self.node_edges.items():
      if node in v:
        self.node_edges[k].remove(node)
    
  def remove_link(self, node1, node2):
    if node1 not in self.nodes or node2 not in self.nodes:
      raise Exception("Одного или более из узлов, связь с которым вы хотите удалить, нет в графе!")

    if not (node2 in self.node_edges[node1] and node1 in self.node_edges[node2]):
      raise Exception("Вы пытаетесь удалить связь, которой нет!")
      
    self.node_edges[node1].remove(node2)
    self.node_edges[node2].remove(node1)
        
      
  
  def print_lists_of_adjacency(self):
    for k, v in self.node_edges.items():
      print(f'{k}: {v}')
  
  @classmethod
  def from_matrix(cls, graph):
    new_graph = GraphOnLists()
    for node in graph.nodes:
      new_graph.add_node(node)
    for i in range(len(graph.matrix)):
      for j in range(i, len(graph.matrix[i])):
        if graph.matrix[i][j] == 1:
          new_graph.add_link(new_graph.nodes[i], new_graph.nodes[j])
    return new_graph
  
  def breadth_first_search(self, value):
    start = self.nodes[0]
    queue = []
    were_here = []
    queue.append(start)
    were_here.append(start)
    current_path = []
    while queue:
      current = queue.pop(0)
      print("Мы здесь - ", current)
      if current.value == value:
      
        return current
      neighbours = self.node_edges[current]
      for i in neighbours:
        if i not in were_here:
          queue.append(i)
          were_here.append(i)
    raise Exception("Нема такого узла!")
  

  def depth_first_search(self, value):
    start = self.nodes[0]
    stos = []
    were_here = []
    stos.append(start)
    while stos:
      current = stos.pop()
      print("Мы здесь - ", current)
      if current not in were_here:
        were_here.append(current)
        if current.value == value:
          return current
        neighbours = self.node_edges[current]
        for i in neighbours:
          stos.append(i)
```

```python
class GraphOnMatrices:
  def __init__(self):
    self.nodes = []
    self.matrix = []
  
  def add_node(self, node):
    if node in self.nodes:
      raise Exception("Вы пытаетесь добавить один и тот же узел дважды!")
    self.nodes.append(node)
    if not self.matrix:
      self.matrix.append([0])
    else:
      for i in range(len(self.matrix)):
        self.matrix[i].append(0)
      self.matrix.append([0 for i in range(len(self.nodes))])
  
  def add_link(self, node1, node2): #Обработать рефлексию!
    if not (node1 in self.nodes and node2 in self.nodes):
      raise Exception("Вы пытаетесь добавить связь с узлами_ом, которого нет в графе!")

    i = self.nodes.index(node1)
    j = self.nodes.index(node2)
    if not(self.matrix[i][j] != 1 and self.matrix[j][i] != 1):
      raise Exception("Вы пытаетесь добавить связь, которая уже есть в графе!")
    self.matrix[i][j] = 1
    self.matrix[j][i] = 1
  
  def remove_link(self, node1, node2):
    if not (node1 in self.nodes and node2 in self.nodes):
      raise Exception("Одного или более из узлов, связь с которым вы хотите удалить, нет в графе!")
    i = self.nodes.index(node1)
    j = self.nodes.index(node2)
    if not (self.matrix[i][j] == 1 and self.matrix[j][i] == 1):
      raise Exception("Вы пытаетесь удалить связь, которой нет!")
    self.matrix[i][j] = 0
    self.matrix[j][i] = 0   
  
  def remove_node(self, node):
    if node not in self.nodes:
      raise Exception("Вы пытаетесь удалить узел, которого нет!")
    i = self.nodes.index(node)
    self.matrix.pop(i)
    self.nodes.pop(i)
    for columns in self.matrix:
      columns.pop(i)     
  
  def print_matrix_of_adjacency(self):
    for i in range(len(self.nodes)):
      print(f'{self.nodes[i]}: {i}')
    
    for i in range(len(self.matrix)):
      print(self.matrix[i])
  
  @classmethod
  def from_lists(cls, graph):
    new_graph = GraphOnMatrices()
    for node in graph.nodes:
      new_graph.add_node(node)
    already_added = []

    for k, v in graph.node_edges.items():
      for node in v:
        if node not in already_added:
          new_graph.add_link(k, node)
      already_added.append(k)
    return new_graph

  def breadth_first_search(self, value):
    start = self.nodes[0]
    queue = []
    were_here = []
    queue.append(start)
    were_here.append(start)
    paths = []
    while queue:
      current = queue.pop(0)
      icurrent = self.nodes.index(current)
      print("Мы здесь - ", current)
      if current.value == value:
        path_current = current
        path = []
        while path_current is not None:
          path.append(path_current)
          path_current = path_current.parent
        path.reverse()
        paths.append(path) 
        were_here.remove(current)
        continue

      neighbours = []
      for j in range(len(self.matrix[icurrent])):
        if self.matrix[icurrent][j] == 1:
          neighbours.append(self.nodes[j])

      for i in neighbours:
        if i not in were_here:
          i.parent = current
          queue.append(i)
          were_here.append(i)
    
    if not paths:
      raise Exception("Нема такого узла!")
    
    return min(paths, key=len)
    
  

  def depth_first_search(self, value):
    start = self.nodes[0]
    stos = []
    were_here = []
    stos.append(start)
    while stos:
      current = stos.pop()
      icurrent = self.nodes.index(current)
      print("Мы здесь - ", current)
      if current not in were_here:
        were_here.append(current)
        if current.value == value:
          return current
        neighbours = []
        for j in range(len(self.matrix[icurrent])):
          if self.matrix[icurrent][j] == 1:
            neighbours.append(self.nodes[j])
        for i in neighbours:
          stos.append(i)
```

```python
class Node:
  def __init__(self, value):
    self.value = value
    self.parent = None
  
  def __hash__(self):
    return hash(self.value)
  
  def __repr__(self):
    return f"Node({self.value.__repr__()})"
  
  def __eq__(self, other):
    return self.value == other.value
```

```python
class GraphOnMatrices:
  def __init__(self):
    self.nodes = []
    self.matrix = []

  def build_chain(self, sentences):
    words = []
    for i in sentences:
      words.append("<start>")
      words += [re.sub(r'[^\w\s]','', i) for i in i.strip().split(" ")]
      words.append("<end>")
    len_matrix = len(set(words))
    for w in set(words):
      self.add_node(Node(w))
    self.statistics(words)

  def statistics(self, words):
    beginning_vector = []
    for i in self.nodes:
      if i.value == "<start>":
        beginning_vector.append(1)
      else:
        beginning_vector.append(0)
    for w1, w2 in zip(words, words[1:]):
      i1 = self.nodes.index(Node(w1))
      i2 = self.nodes.index(Node(w2))
      self.matrix[i1][i2] += 1
    
    for i in range(len(self.matrix)):
      for j in range(len(self.matrix)):
        self.matrix[i][j]/=len(self.nodes)

  def generate_random_string(self, start_word="<start>"):
    print(start_word, end = " ")
    while True:
      i = self.nodes.index(Node(start_word))
      weights = self.matrix[i]
      start_word = random.choices(self.nodes, weights = weights, k = 1)[0].value
      print(start_word, end = " ")
      if start_word == "<end>":
        print("")
        break
    
        

  def add_node(self, node):
    if node in self.nodes:
      raise Exception("Вы пытаетесь добавить один и тот же узел дважды!")
    self.nodes.append(node)
    if not self.matrix:
      self.matrix.append([0])
    else:
      for i in range(len(self.matrix)):
        self.matrix[i].append(0)
      self.matrix.append([0 for i in range(len(self.nodes))])
  
  def add_link(self, node1, node2): #Обработать рефлексию!
    if not (node1 in self.nodes and node2 in self.nodes):
      raise Exception("Вы пытаетесь добавить связь с узлами_ом, которого нет в графе!")

    i = self.nodes.index(node1)
    j = self.nodes.index(node2)
    if not(self.matrix[i][j] != 1 and self.matrix[j][i] != 1):
      raise Exception("Вы пытаетесь добавить связь, которая уже есть в графе!")
    self.matrix[i][j] = 1
    self.matrix[j][i] = 1
  
  def remove_link(self, node1, node2):
    if not (node1 in self.nodes and node2 in self.nodes):
      raise Exception("Одного или более из узлов, связь с которым вы хотите удалить, нет в графе!")
    i = self.nodes.index(node1)
    j = self.nodes.index(node2)
    if not (self.matrix[i][j] == 1 and self.matrix[j][i] == 1):
      raise Exception("Вы пытаетесь удалить связь, которой нет!")
    self.matrix[i][j] = 0
    self.matrix[j][i] = 0   
  
  def remove_node(self, node):
    if node not in self.nodes:
      raise Exception("Вы пытаетесь удалить узел, которого нет!")
    i = self.nodes.index(node)
    self.matrix.pop(i)
    self.nodes.pop(i)
    for columns in self.matrix:
      columns.pop(i)     
  
  def print_matrix_of_adjacency(self):
    for i in range(len(self.nodes)):
      print(f'{self.nodes[i]}: {i}')
    
    for i in range(len(self.matrix)):
      print(self.matrix[i])
  
  @classmethod
  def from_lists(cls, graph):
    new_graph = GraphOnMatrices()
    for node in graph.nodes:
      new_graph.add_node(node)
    already_added = []

    for k, v in graph.node_edges.items():
      for node in v:
        if node not in already_added:
          new_graph.add_link(k, node)
      already_added.append(k)
    return new_graph

  def breadth_first_search(self, value):
    start = self.nodes[0]
    queue = []
    were_here = []
    queue.append(start)
    were_here.append(start)
    paths = []
    while queue:
      current = queue.pop(0)
      icurrent = self.nodes.index(current)
      print("Мы здесь - ", current)
      if current.value == value:
        path_current = current
        path = []
        while path_current is not None:
          path.append(path_current)
          path_current = path_current.parent
        path.reverse()
        paths.append(path) 
        were_here.remove(current)
        continue

      neighbours = []
      for j in range(len(self.matrix[icurrent])):
        if self.matrix[icurrent][j] == 1:
          neighbours.append(self.nodes[j])

      for i in neighbours:
        if i not in were_here:
          i.parent = current
          queue.append(i)
          were_here.append(i)
    
    if not paths:
      raise Exception("Нема такого узла!")
    
    return min(paths, key=len)
    
  

  def depth_first_search(self, value):
    start = self.nodes[0]
    stos = []
    were_here = []
    stos.append(start)
    while stos:
      current = stos.pop()
      icurrent = self.nodes.index(current)
      print("Мы здесь - ", current)
      if current not in were_here:
        were_here.append(current)
        if current.value == value:
          return current
        neighbours = []
        for j in range(len(self.matrix[icurrent])):
          if self.matrix[icurrent][j] == 1:
            neighbours.append(self.nodes[j])
        for i in neighbours:
          stos.append(i)
```

```python
def preprocess(texts):
  preprocessed_texts = []
  for text in texts:
    correctedstring = " ".join(text.split()) 
    correctedstring = correctedstring.lower()
    preprocessed_texts += [i.strip() for i in re.split(r"[;?.!]+\s", correctedstring) if len(i) != 0]
  return preprocessed_texts
```
